```python
movies=movie.iloc[:, 1:2]

movies['Movie_Title']= movies['Movie_Title'].str.strip()
movies
```

## Count Vectorizer

## Getting feature text conversion into Tokens

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
tfidf_matrix = tf.fit_transform(movies['Movie_Title'])
print(tfidf_matrix.shape)

from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
print(cosine_sim.shape)


# Build a 1-dimensional array with movie titles
titles = movies['Movie_Title']
indices = pd.Series(movies.index, index=movies['Movie_Title'])
```

## **Modeling**

```python
print(cosine_sim)
```

## Getting movie name from user and validate closest spelling
****
Printing top 30 closest movie titles

```python
title=input("Enter the movie related to recommend:")
num=30

idx = indices[title]

sim_scores = list(enumerate(cosine_sim[idx]))
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
sim_scores = sim_scores[1:num+1]
movie_indices = [i[0] for i in sim_scores]
   
scores=[i[1] for i in sim_scores]

print("Recommending 30 movies similar to " + title + "...")   
print("-------")    
for rec in range(num): 
    print("Recommended: " + titles[movie_indices[rec]] + " (score:" +      str(scores[rec]) + ")")
```

## **Model Evaluation**

```python
title=input("Enter the movie related to recommend:")
num=10

idx = indices[title]

sim_scores = list(enumerate(cosine_sim[idx]))
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
sim_scores = sim_scores[1:num+1]
movie_indices = [i[0] for i in sim_scores]
   
scores=[i[1] for i in sim_scores]

print("Recommending 10 movies similar to " + title + "...")   
print("-------")    
for rec in range(num): 
    print("Recommended: " + titles[movie_indices[rec]] + " (score:" +      str(scores[rec]) + ")")
```
